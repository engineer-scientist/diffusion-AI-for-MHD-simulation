```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
avg_MSE_list = [0.7296, 0.4908, 0.4198, 0.4006, 0.3824]
avg_div_B_list = [232.7913, 45.1533, 21.7838, 12.1204, 7.8321]
```

```python
import matplotlib.ticker as ticker

fig, ax1 = plt.subplots()

# Define x-axis values from 1 to 5
epochs = np.arange(1, len(avg_MSE_list) + 1)

# Plot MSE on the left axis
color1 = 'tab:blue'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Average mean square error', color=color1)
ax1.plot(epochs, avg_MSE_list, 'o-', color=color1, label='MSE')
ax1.tick_params(axis='y', labelcolor=color1)

# Create a second y-axis for div_B
ax2 = ax1.twinx()
color2 = 'tab:orange'
ax2.set_ylabel('Average divergence of magnetic field B', color=color2)
ax2.plot(epochs, avg_div_B_list, 's-', color=color2, label='div_B')
ax2.tick_params(axis='y', labelcolor=color2)

# Scale axis limits: 300 units of div_B = 1 unit of MSE
ax1_max = max(avg_MSE_list) * 1.2
ax1.set_ylim(0, ax1_max)
ax2.set_ylim(0, ax1_max * 300)

# Expand X-axis from 0 to 6 to make all points clearly visible
ax1.set_xlim(0.5, 5.5)
# Ensure integer ticks are used
ax1.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.grid(True)
plt.savefig('MSE and div B training plot')#, bbox_inches = 'Tight')
```
